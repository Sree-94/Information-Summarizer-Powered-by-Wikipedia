# Generating extractive summary

```python
import nltk
nltk.download('punkt')
sentence_list = nltk.sent_tokenize(article_text)
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')

word_frequencies = {}
for word in nltk.word_tokenize(formatted_article_text):
    if word not in stopwords:
        if word not in word_frequencies.keys():
            word_frequencies[word] = 1
        else:
            word_frequencies[word] += 1
```

```python
maximum_frequncy = max(word_frequencies.values())

for word in word_frequencies.keys():
    word_frequencies[word] = (word_frequencies[word]/maximum_frequncy)
```

```python
individual_sentence_scores = {}
for each_sentence in sentence_list:
    for word in nltk.word_tokenize(each_sentence.lower()):
        if word in word_frequencies.keys():
            if len(each_sentence.split(' ')) < 30:
                if each_sentence not in individual_sentence_scores.keys():
                    individual_sentence_scores[each_sentence] = word_frequencies[word]
                else:
                    individual_sentence_scores[each_sentence] += word_frequencies[word]
```

```python
import heapq
extractive_summary_sentences = heapq.nlargest(number_of_headings, individual_sentence_scores, key=individual_sentence_scores.get)

extractive_summary = ' '.join(extractive_summary_sentences)
print(extractive_summary)
html_summarypage +=(extractive_summary+slice3)
```

```python
def abstractive_summary_pretrained_function(article_data):
  # Create tokens - number representation of our text
  ab1_tokens = ab1_tokenizer(article_data, truncation=True, padding="longest", return_tensors="pt")

  # Summarize 
  summary_ab1 = model_pretrained.generate(**ab1_tokens)

  # Decode summary
  abstractive_summary_pretrained=ab1_tokenizer.decode(summary_ab1[0])

  return abstractive_summary_pretrained
#print(abstractive_summary_pretrained)
```

```python
def flatten(example):
    return {
        "document": example["document"],#["document"],
        "summary": example["summary"],#["summary"],
    }


dataset = data.map(flatten, remove_columns=["document", "summary", "id"])
train_data_txt, validation_data_txt = dataset.train_test_split(test_size=0.1).values()
```

```python
def batch_tokenize_preprocess(batch, tokenizer, max_source_length, max_target_length):
    source, target = batch["document"], batch["summary"]
    source_tokenized = tokenizer(
        source, padding="max_length", truncation=True, max_length=max_source_length
    )
    target_tokenized = tokenizer(
        target, padding="max_length", truncation=True, max_length=max_target_length
    )

    batch = {k: v for k, v in source_tokenized.items()}
    # Ignore padding in the loss
    batch["labels"] = [
        [-100 if token == tokenizer.pad_token_id else token for token in l]
        for l in target_tokenized["input_ids"]
    ]
    return batch


train_data = train_data_txt.map(
    lambda batch: batch_tokenize_preprocess(
        batch, tokenizer, encoder_max_length, decoder_max_length
    ),
    batched=True,
    remove_columns=train_data_txt.column_names,
)

validation_data = validation_data_txt.map(
    lambda batch: batch_tokenize_preprocess(
        batch, tokenizer, encoder_max_length, decoder_max_length
    ),
    batched=True,
    remove_columns=validation_data_txt.column_names,
)
```

```python

nltk.download("punkt", quiet=True)

metric = datasets.load_metric("rouge")


def postprocess_text(preds, labels):
    preds = [pred.strip() for pred in preds]
    labels = [label.strip() for label in labels]

    # rougeLSum expects newline after each sentence
    preds = ["\n".join(nltk.sent_tokenize(pred)) for pred in preds]
    labels = ["\n".join(nltk.sent_tokenize(label)) for label in labels]

    return preds, labels


def compute_metrics(eval_preds):
    preds, labels = eval_preds
    if isinstance(preds, tuple):
        preds = preds[0]
    decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)
    # Replace -100 in the labels as we can't decode them.
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # Some simple post-processing
    decoded_preds, decoded_labels = postprocess_text(decoded_preds, decoded_labels)

    result = metric.compute(
        predictions=decoded_preds, references=decoded_labels, use_stemmer=True
    )
    # Extract a few results from ROUGE
    result = {key: value.mid.fmeasure * 100 for key, value in result.items()}

    prediction_lens = [
        np.count_nonzero(pred != tokenizer.pad_token_id) for pred in preds
    ]
    result["gen_len"] = np.mean(prediction_lens)
    result = {k: round(v, 4) for k, v in result.items()}
    return result
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir="results",
    num_train_epochs=1,  # demo
    do_train=True,
    do_eval=True,
    per_device_train_batch_size=4,  # demo
    per_device_eval_batch_size=4,
    # learning_rate=3e-05,
    warmup_steps=500,
    weight_decay=0.1,
    label_smoothing_factor=0.1,
    predict_with_generate=True,
    logging_dir="logs",
    logging_steps=50,
    save_total_limit=3,
)

data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_data,
    eval_dataset=validation_data,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
if WANDB_INTEGRATION:
    wandb_run = wandb.init(
        project="bart_cnn_dailymail",
        config={
            "per_device_train_batch_size": training_args.per_device_train_batch_size,
            "learning_rate": training_args.learning_rate,
            "dataset": "cnn_dailymail ",
        },
    )

    now = datetime.now()
    current_time = now.strftime("%H%M%S")
    wandb_run.name = "run_" + "_" + current_time
```

```python
#%%wandb
# uncomment to display Wandb charts

trainer.train()
```

```python
def abstractive_summary_generator(article_content, model):
    inputs = tokenizer(
        article_content,
        padding="max_length",
        truncation=True,
        max_length=encoder_max_length,
        return_tensors="pt",
    )
    input_ids = inputs.input_ids.to(model.device)
    attention_mask = inputs.attention_mask.to(model.device)
    outputs = model.generate(input_ids, attention_mask=attention_mask)
    output_str = tokenizer.batch_decode(outputs, skip_special_tokens=True)
    return output_str

test_samples = validation_data_txt.select(range(16))

print(abstractive_summary_generator(article_text, model))
```

```python
# natural language processing: n-gram ranking
import re
import unicodedata
import nltk
from nltk.corpus import stopwords# add appropriate words that will be ignored in the analysis
```

```python
def clean_textdata(text):
  wnl = nltk.stem.WordNetLemmatizer()
  stopwords = nltk.corpus.stopwords.words('english')
  text = (unicodedata.normalize('NFKD', text)
    .encode('ascii', 'ignore')
    .decode('utf-8', 'ignore')
    .lower())
  words = re.sub(r'[^\w\s]', '', text).split()
  return [wnl.lemmatize(word) for word in words if word not in stopwords]
```

# Stacked Barplot

```python
common_words_list=[]
common_words_tupple=list(set((list(set(list(unigrams_series_extractive.index)).intersection(list(unigrams_series_abstractive.index))))).intersection(list(unigrams_series_article.index)))
for i in range(0,len(common_words_tupple)):
  common_words_list.append(list(common_words_tupple[i])[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt


def stacked_plot(results, label_names):
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    label_shades = plt.get_cmap('bwr_r')(
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(label_names, label_shades)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        ax.barh(labels, widths, left=starts, height=0.5,
                label=colname, color=color)
        xcenters = starts + widths / 2

        r, g, b, _ = color
        text_color = 'black'
        for y, (x, c) in enumerate(zip(xcenters, widths)):
            ax.text(x, y, str(int(c)), ha='center', va='center',
                    color=text_color)
    ax.legend(ncol=len(label_names), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='small')

    return fig, ax


label_names = ['Orginal Article', 'Extractive Summary','Abstractive Summary']

words_dict = {}

for i in range(0,len(common_words_list)):
  words_dict[common_words_list[i]]=[unigrams_series_article[common_words_tupple[i]],
                                 unigrams_series_extractive[common_words_tupple[i]],
                                 unigrams_series_abstractive[common_words_tupple[i]]]

print(words_dict)

stacked_plot(words_dict, label_names)


tmpfile = BytesIO()
plt.savefig(tmpfile)
encoded = base64.b64encode(tmpfile.getvalue()).decode('utf-8')
html = '<img src=\'data:image/png;base64,{}\'>'.format(encoded)
print(html)
html_summarypage += html+slice14

plt.show()
```

# Lollipop Plot

```python
import pandas as pd
import numpy as np
```

```python
def lollipop_plot(Function_data, plot_title,slice_1):
  global html_summarypage
  resultant_df=Function_data.to_frame(name="count").reset_index().drop(['index'], axis=1)
  temp_row1=[]
  for i in range(0,len(Function_data)):
    temp_row1.append(list(list(Function_data.index)[i])[0])

  resultant_df['words'] = temp_row1
  resultant_df.head()
  my_range=range(1,len(resultant_df.index)+1)
  # Make the plot
  plt.figure(figsize=(10, 5))
  plt.title(plot_title)
  plt.ylabel("Count")
  plt.stem(resultant_df['count'])
  plt.xticks( my_range, resultant_df['words'])
  plt.xticks(my_range, rotation='vertical')

  tmpfile = BytesIO()
  plt.savefig(tmpfile)
  encoded = base64.b64encode(tmpfile.getvalue()).decode('utf-8')
  html = '<img src=\'data:image/png;base64,{}\'>'.format(encoded)
  print(html)
  html_summarypage += html+slice_1
  return 0
```
